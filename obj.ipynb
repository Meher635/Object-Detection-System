```python
import cv2
from ultralytics import YOLO
import pyttsx3
import os
import time

# ------------------ VOICE SETUP ------------------
engine = pyttsx3.init()
engine.setProperty('rate', 150)

# ------------------ MODEL LOAD ------------------
model = YOLO("yolov8n.pt")

# ------------------ MENU ------------------
print("1. Image Detection")
print("2. Real-Time Camera Detection")
print("3. Exit")

choice = input("Enter choice (1/2/3): ")

# ================= IMAGE =================
if choice == "1":
    path = input("Enter image path: ")

    if os.path.isfile(path):
        img = cv2.imread(path)

        if img is None:
            print("Image load nahi hui")
        else:
            results = model(img)
            output = results[0].plot()

            names = results[0].names
            boxes = results[0].boxes

            detected_objects = set()

            for box in boxes:
                cls_id = int(box.cls[0])
                detected_objects.add(names[cls_id])

            if detected_objects:
                text = "Detected " + ", ".join(sorted(detected_objects))
            else:
                text = "No object detected"

            print(text)
            engine.say(text)
            engine.runAndWait()

            cv2.imshow("Image Detection", output)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
    else:
        print("Invalid path")

# ================= CAMERA =================
elif choice == "2":
    cap = cv2.VideoCapture(0)

    if not cap.isOpened():
        print("Camera open nahi hua")
    else:
        last_objects = set()   # 🔥 important
        last_spoken_time = 0
        cooldown = 3

        import threading
        def speak(text):
            def run():
                engine.say(text)
                engine.runAndWait()
            threading.Thread(target=run).start()

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            results = model(frame)
            output = results[0].plot()

            names = results[0].names
            boxes = results[0].boxes

            current_objects = set()

            for box in boxes:
                cls_id = int(box.cls[0])
                current_objects.add(names[cls_id])

            # ✅ Stable text
            if current_objects:
                text = "Detected " + ", ".join(sorted(current_objects))
            else:
                text = "No object detected"

            current_time = time.time()

            # 🔥 REAL FIX (object-level comparison)
            if (
                current_objects != last_objects and
                (current_time - last_spoken_time > cooldown)
            ):
                print(text)
                speak(text)

                last_objects = current_objects.copy()
                last_spoken_time = current_time

            cv2.imshow("Real-Time Detection", output)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        cap.release()
        cv2.destroyAllWindows()

# ================= EXIT =================
elif choice == "3":
    print("Exit")

else:
    print("Wrong choice") 
```
